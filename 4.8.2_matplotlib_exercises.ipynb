```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np
```

```python
x = range(1, 25)
y = [n ** 2 - n + 2 for n in x]

plt.annotate("Origin", xy=(0,0), xytext=(0, 125), arrowprops={'facecolor': 'blue'})
plt.plot(y)
plt.show()
```

```python
x = range(100)
y = [n ** (1/2) for n in x]

plt.plot(y)
```

```python
x = range(20)
y = [n ** 3 for n in x]

plt.plot(x, y)
```

```python
x = np.arange(-math.pi, math.pi, .1)
y = [math.tan(n) for n in x]

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-5, 5)
```

```python
x = range(-10, 11)
y = [2 ** n for n in x]

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
```
