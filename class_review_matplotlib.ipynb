```python
%matplotlib inline
import matplotlib.pyplot as plt
import math
import numpy as np
```

## Exercise 1
* Plot $y=x^2 -x + 2x$
* Write code that generates the x and y points
* Annotate the origin at (0, 0)

```python
# y = x^2 - x + 2x

x = range(1, 25)
y = [n ** 2 - n + 2 for n in x]

plt.title("$y=x^2 -x +2x$")
plt.xlabel("x")
plt.ylabel("y", rotation=0)

plt.annotate("Origin", xy=(0,0), xytext=(0,75), arrowprops={'facecolor': 'green'})
plt.plot(y)
plt.show()
```

## Exercise 2

* Plot $y = \sqrt{x}$
* Plot $y = x^3$
* Plot $y = tan(x)$
* Plot $y = 2^x$

```python
# y = sqrt(x)

x = range(1, 100)
y = [math.sqrt(n) for n in x]

plt.title("$y = \sqrt{x}$", fontsize=18)

# set the fontsize
plt.xlabel("x", fontsize=14)

# y label with rotation set to 0 degrees
plt.ylabel("$\sqrt{x}$", rotation=0, fontsize=14)

plt.plot(y)
plt.show()
```

```python
# y = x^3

x = list(range(1, 100))
y = [n**3 for n in x]

plt.title("$y = x^3$", fontsize=14)

plt.xlabel('x', fontsize=14)
plt.ylabel("$x^3$", rotation=0, fontsize=14)

# plt.yscale("log")
plt.plot(y)
plt.show()
```

```python
# .linspace arguments are (start, end, number_of_steps)
x = np.linspace(-2 * math.pi, 2 * math.pi, 1000)

# np.tan(x) is similar to [math.tan(n) for n in x]
y = np.tan(x)

# this operation inserts a NaN where the difference between successive points is negative
# NaN means "Not a Number" and NaNs are not plotted or connected
# I found this by doing a search for "How to plot tan(x) in matplotlib without the connecting lines between asymtotes"
y[:-1][np.diff(y) < 0] = np.nan

# show grid
plt.grid()

plt.xlabel("x")
plt.ylabel("$tan(x)")

# set the x and y axis cutoffs
plt.ylim(-10, 10)
plt.xlim(-2 * math.pi, 2 * math.pi)

# x_labels in radians
# For more programmatic approach to radians, see https://matplotlib.org/3.1.1/gallery/units/radian_demo.html
radian_multiples = [-2, -3/2, -1, -1/2, 0, 1/2, 1, 3/2, 2]
radians = [n * math.pi for n in radian_multiples]
radian_labels = ['$-2\pi$', '$-3\pi/2$', '$\pi$', '$-\pi/2$', '0', '$\pi/2$', '$\pi$', '$3\pi/2$', '$2\pi$']

plt.xticks(radians, radian_labels)

plt.title("$y = tan(x)$", fontsize=14)
plt.plot(x, y)
plt.show()
```

```python
x = range(1, 40)
y = [2**n for n in x]

plt.title("$y = 2^x$")
plt.xlabel("x")
plt.ylabel("$y= 2^x$")

# plt.yscale("log")
plt.plot(y)
plt.show()
```

## Exercise 3

Combine the last 4 figures into one large figure with 4 subplots.

```python
plt.figure(figsize=(16, 6))
plt.suptitle('Subplots for the above 4 exercises', fontsize=16)
plt.subplots_adjust(hspace=0.5)

# first subplot
# y = x^2 + x + x2
plt.subplot(2, 2, 1)

x = range(1, 50)
y = [math.sqrt(n) for n in x]

plt.title("$y=\sqrt{x}$")
plt.xlabel("x")
plt.ylabel("y", rotation=0)

plt.plot(y)

# second subplot
# y = x^3
plt.subplot(2, 2, 2)

y = [n**3 for n in x]
plt.title("$y = x^3$")
plt.ylabel("$x^3$", rotation=0)
plt.plot(y)

# third subplot
# y = tan(x)

plt.subplot(2, 2, 3)

x = np.linspace(-2 * np.pi, 2 * np.pi, 500)
y = np.tan(x)

# set NaN for when we go from -infinity to +infinity asymtotes
y[:-1][np.diff(y) < 0] = np.nan

plt.grid()
plt.ylim(-10, 10)
plt.xlim(-2 * np.pi, 2 * np.pi)

plt.title("$y=tan(x)$")
plt.xlabel("x")
plt.ylabel("y", rotation=0)
plt.plot(x, y)

# fourth subplot
# y = 2^x
plt.subplot(2, 2, 4)

x = range(1, 15)
y = [2**n for n in x]

plt.title("$y=2^x$")
plt.xlabel("x")
plt.ylabel("$y=2^x$")
plt.plot(y)

plt.show
```

## Exercise 4 - plot the above 4 curves on a single chart with different colored lines

Be sure to include a legend

```python
x = np.linspace(0, 15, 1000)

# y = sqrt(x)
ex1 = [math.sqrt(n) for n in x]

# y = x ^ 3
ex2 = [n**3 for n in x]

# y = tan(x)
ex3 = [math.tan(n) for n in x]
ex3 = np.array(ex3)
ex3[:-1][np.diff(ex3) < 0] = np.nan

# y = 2 ^ x
ex4 = [2**n for n in x]

plt.figure(figsize=(16, 9))

plt.ylim(-25, 25)
plt.xlim(0, 12)
plt.plot(x, ex1, c='blue', label='$\sqrt{x}$')
plt.plot(x, ex2, c='orange', label='$x^3$')
plt.plot(x, ex3, c='green', label='$tan(x)$')
plt.plot(x, ex4, c='purple', label='$2^x$')

plt.legend(loc='best')
plt.title('4 Different Curves on One Graph')
plt.xlabel('$x$')
plt.grid()
plt.show()
```
